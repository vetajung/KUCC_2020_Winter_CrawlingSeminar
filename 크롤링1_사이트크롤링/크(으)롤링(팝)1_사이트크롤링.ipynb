```python
from bs4 import BeautifulSoup #html의 태그를 다룰 수 있는 라이브러리
from urllib.request import urlopen #urllib 는 URL 작업을 위한 여러 모듈을 모은 패키지
import re # regular expression, 텍스트 처리할때 사용
```

re 모듈은 매우 쓸모로움(대충많이 쓸모있다는 뜻), 다음 링크를 통해 필요시 공부하기: https://whatisthenext.tistory.com/116

크롤링 할 사이트: https://www.greatplacetowork.com/best-workplaces/100-best/2019

```python
# 2019 Fortune 100 Best Companies to Work For
url  = 'https://www.greatplacetowork.com/best-workplaces/100-best/2019'
html = urlopen(url)
soup = BeautifulSoup(html, "lxml")
```

![title](website.png)

우리는 위의 사이트에서 순위, 기업명, 업종, 지역을 크롤링해서 엑셀로 저장하는게 목표임

### 수집하고자 하는 데이터를 class명을 이용해서 구체적으로 찾아보기

```python
# 각 회사당 정보를 포함하는 전체 셀을 찾아 class 이름 찾기
listOfsoup = soup.find_all('div', 'col-md-5 col-xs-12 company-text')
print(listOfsoup)
```

```python
# 총 몇개인가 확인해보니 100개 기업 리스트에 맞게 100개가 나왔음
len(listOfsoup)
```

```python
# 링크,기업명에 해당하는 부분을 찾음, (a class)
listOfsoup[0].find("a")
```

```python
# 100개의 기업에 대해 리스트 생성
rank = []
company = []
industry = []
location = []
url = []

for item in listOfsoup:
    rank.append(item.find("div","rank large").get_text().strip())
    company.append(item.find("a")['title'])
    industry.append(item.find("li").get_text().strip())
    location.append(item.find_all("li")[1].get_text())
    url.append(item.find("a")['href'])
```

### 데이터를 데이터프레임으로 정리 후, 파일로 저장

```python
# 데이터프레임 생성
import pandas as pd

data = {'Rank':rank, 'Company':company, 'Industry':industry, 'Location':location, 'URL':url}
df = pd.DataFrame(data)
df.head(10)
```

```python
# 파일로 저장
df.to_csv('100_Best_Companies_to_Work_For.csv', sep=',', encoding='UTF-8',index = False)
```
